```python
# prompt: Implement ArxivAPIWrapper to fetch research papers and WikipediaAPIWrapper to retrieve general knowledge. Ensure both tools run independently and return relevant results.

import requests

class ArxivAPIWrapper:
    def __init__(self):
        self.base_url = "http://export.arxiv.org/api/query?"

    def fetch_papers(self, query, max_results=10):
        params = {
            "search_query": query,
            "max_results": max_results,
            "sortBy": "relevance",
            "sortOrder": "descending"
        }

        response = requests.get(self.base_url, params=params)

        if response.status_code == 200:
            # Process the response (XML) to extract relevant paper information here.
            # You may use libraries like 'feedparser' to simplify XML parsing.
            return response.text  # Return the raw XML for now.
        else:
            return f"Error fetching data: {response.status_code}"

class WikipediaAPIWrapper:
    def __init__(self):
        self.base_url = "https://en.wikipedia.org/w/api.php"

    def fetch_summary(self, query):
        params = {
            "action": "query",
            "format": "json",
            "titles": query,
            "prop": "extracts",
            "exintro": True,
            "explaintext": True
        }

        response = requests.get(self.base_url, params=params)

        if response.status_code == 200:
            data = response.json()
            pages = data["query"]["pages"]
            for page_id, page_data in pages.items():
                if "extract" in page_data:
                    return page_data["extract"]
        else:
            return f"Error fetching data: {response.status_code}"


# Initialize the Arxiv tool
arxiv_tool = ArxivAPIWrapper()

# Fetch research papers related to AI
ai_papers = arxiv_tool.fetch_papers("artificial intelligence")
print("Arxiv Results:\n", ai_papers)

# Initialize the Wikipedia tool
wiki_tool = WikipediaAPIWrapper()

# Fetch general information from Wikipedia
wiki_summary = wiki_tool.fetch_summary("Artificial intelligence")
print("\nWikipedia Results:\n", wiki_summary)
```
