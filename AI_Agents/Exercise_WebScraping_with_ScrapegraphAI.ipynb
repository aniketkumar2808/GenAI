### Exercise 1: E-commerce Price Comparison
Build a multi-site scraper that compares prices for a specific product across different e-commerce platforms.


#### Hint

```python
import json
from scrapegraphai.graphs import SmartScraperMultiGraph
# Configure multi-scraper

multiple_search_graph = SmartScraperMultiGraph(
    prompt="Fetch the prices of track pant for men",
    source= [
        "https://www.myntra.com/yoga-track-pants-men?extra_search_param=isautosuggestentry%3atrue%3a%3aid%3a2297-yoga-track-pants-men&rawQuery=Yoga%20Track%20Pants%20Men",
        "https://www.amazon.in/s?k=yoga+pants+for+men&crid=1D8XR677REG8E&sprefix=Yoga+Track+Pants+Men%2Caps%2C235&ref=nb_sb_ss_ts-doa-p_1_20"
        ],
    schema=None,
    config=graph_config_openai
)

# Run comparison
result = multiple_search_graph.run()
print(json.dumps(result, indent=4))
```
